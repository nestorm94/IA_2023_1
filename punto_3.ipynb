```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Cargar los datos
data = pd.read_csv('boston_housing.csv')

# Seleccionar las características y el objetivo
features = data.drop('medv', axis=1)
target = data['medv']

# Dividir los datos en un conjunto de entrenamiento y un conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Ajustar el modelo de árbol de decisión
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X_train, y_train)

# Ajustar el modelo de bosques aleatorios
forest_model = RandomForestRegressor(random_state=42)
forest_model.fit(X_train, y_train)

# Hacer predicciones y calcular el error cuadrado medio y el coeficiente de determinación
for model in [tree_model, forest_model]:
    predictions = model.predict(X_test)
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    print(f'Modelo: {model.__class__.__name__}')
    print(f'Media Absoluta de Error: {mae:.2f}')
    print(f'R^2 Score: {r2:.2f}\n')
```

```python
import matplotlib.pyplot as plt

for model in [tree_model, forest_model]:
    predictions = model.predict(X_test)
    residuals = y_test - predictions
    plt.figure(figsize=(10,6))
    plt.scatter(y_test, residuals)
    plt.axhline(y=0, color='r', linestyle='-')
    plt.title(f'Residuos para  {model.__class__.__name__}')
    plt.xlabel('Observacion')
    plt.ylabel('Residuos')
    plt.show()

```
